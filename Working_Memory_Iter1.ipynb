#First Iteration of Working Memory Model
##Model
###states
* s_1: stm1, M(stm1)   (stimulus 1 is presented, stimulus 1's correlation is in memory)
* s_2: stm1, M(stm2)
* s_3: stm2, M(stm1)
* s_4: stm2, M(stm2)

###rewards
* s_1: 1
* s_2: 0
* s_3: 0
* s_4: 1

###actions
* a_1: replace memory with recently presented correlation
* a_2: keep memory

###environment
* alpha: encodes probability of presented stimulus switching between time steps.


```python
import wm_iter1
from frozen_dict import *
import matplotlib.pyplot as plt
import pdb

%matplotlib inline

alpha = 0.0
alpha_list = list()
q_s1_replace = list()
q_s1_ignore = list()

q_s2_replace = list()
q_s2_ignore = list()

q_s3_replace = list()
q_s3_ignore = list()

q_s4_replace = list()
q_s4_ignore = list()

q_s5_replace = list()
q_s5_ignore = list()

q_s6_replace = list()
q_s6_ignore = list()

while alpha <= 1.0:
    q, q_table = wm_iter1.main(alpha)
    
    q_s1_replace.append(q[frozendict({'stimulus': 'stim1', 'wm': ('stim1',)}), 'replace'])
    q_s1_ignore.append(q[frozendict({'stimulus': 'stim1', 'wm': ('stim1',)}), 'ignore'])
    
    q_s2_replace.append(q[frozendict({'stimulus': 'stim1', 'wm': ('stim2',)}), 'replace'])
    q_s2_ignore.append(q[frozendict({'stimulus': 'stim1', 'wm': ('stim2',)}), 'ignore'])
    
    q_s3_replace.append(q[frozendict({'stimulus': 'stim2', 'wm': ('stim1',)}), 'replace'])
    q_s3_ignore.append(q[frozendict({'stimulus': 'stim2', 'wm': ('stim1',)}), 'ignore'])
    
    q_s4_replace.append(q[frozendict({'stimulus': 'stim2', 'wm': ('stim2',)}), 'replace'])
    q_s4_ignore.append(q[frozendict({'stimulus': 'stim2', 'wm': ('stim2',)}), 'ignore'])
    
    q_s5_replace.append(q[frozendict({'stimulus': 'stim1', 'wm': ()}), 'replace'])
    q_s5_ignore.append(q[frozendict({'stimulus': 'stim1', 'wm': ()}), 'ignore'])
    
    q_s6_replace.append(q[frozendict({'stimulus': 'stim2', 'wm': ()}), 'replace'])
    q_s6_ignore.append(q[frozendict({'stimulus': 'stim2', 'wm': ()}), 'ignore'])
    alpha_list.append(alpha)
    alpha += .05

```

```python
import numpy

#print len(alpha)
#print len(q_s1)

#plt.text(.05,2.3, 'stimulus switches')
#plt.text(.8, 2.3, 'stimulus stays')
q_s1 = numpy.subtract(q_s1_replace,q_s1_ignore)
q_s2 = numpy.subtract(q_s2_replace,q_s2_ignore)
q_s3 = numpy.subtract(q_s3_replace,q_s3_ignore)
q_s4 = numpy.subtract(q_s4_replace,q_s4_ignore)
q_s5 = numpy.subtract(q_s5_replace,q_s5_ignore)
q_s6 = numpy.subtract(q_s6_replace,q_s6_ignore)

plt.figure(1)
p1, = plt.plot(alpha_list, q_s1, '-', label="stim 1 shown, stim 1 in memory")
p4, = plt.plot(alpha_list, q_s4, '-', label="stim2 shown, stim 2 in memory")
zero, = plt.plot(alpha_list, numpy.zeros(len(alpha_list)), ':')
l1 = plt.legend([p1,p4,zero], ["stim 1 shown, stim 1 in memory", "stim2 shown, stim 2 in memory"], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.ylabel('Q(s,replace) - Q(s,ignore)')
plt.xlabel('Likelihood of Stimulus Switching')
plt.title('Correct Stimulus in memory')
plt.show

plt.figure(2)
p2, = plt.plot(alpha_list, q_s2, '-', label="stim1 shown, stim 2 in memory")
p3, = plt.plot(alpha_list, q_s3, '-', label="stim2 shown, stim 1 in memory")
zero, = plt.plot(alpha_list, numpy.zeros(len(alpha_list)), ':')
l2 = plt.legend([p2,p3,zero], ["stim1 shown, stim 2 in memory", "stim2 shown, stim 1 in memory"], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Wrong stimulus in memory')
plt.ylabel('Q(s,replace) - Q(s,ignore)')
plt.xlabel('Likelihood of Stimulus Switching')
plt.show

plt.figure(3)
p5, = plt.plot(alpha_list,q_s5, '-', label="stim1 shown, empty memory")
p6, = plt.plot(alpha_list,q_s6, '-', label="stim2 shown, empty memory")
zero, = plt.plot(alpha_list, numpy.zeros(len(alpha_list)), ':')
l3 = plt.legend([p5,p6,zero], ["stim1 shown, empty memory", "stim2 shown, empty memory"], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show
plt.title('Empty Memory')
plt.ylabel('Q(s,replace) - Q(s,ignore)')
plt.xlabel('Likelihood of Stimulus Switching')

    
#m1, = plt.plot
```

When the stimulus in memory is the same as the stimulus shown, replacing the contents of memory has the same effect as maintaining the prior contents.  Thus, regardless of the likelihood of the stimulus to switch between trials, there is no benefit to favoring replacing or ignoring, as reflected in the small discrepancy in Q-values between ignoring and replacing.When the stimulus in memory does not match the stimulus presented it tends to be a better strategy to replace memories when the stimulus presented is not likely to switch.  When the presented stimulus is likely to switch, the best that the agent can do is to retain the current memory.  When the stimulus never changes, it is impossible for another stimulus to ever be stored in memory.  Because this state is never reached, both Q-values at this state remain at their default values, 0.  When there is no stimulus in memory it is best to replace memory because the curent contents will never generate immediate reward.  When the stimulus presented is likely to switch, the agent will have a delayed reward, making the Q-value of replacing at that timestep lower than if the stimulus were likely to remain the same.
