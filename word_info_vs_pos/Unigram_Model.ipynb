```python
%matplotlib inline

data = list(csv.reader(open("pos_info_3_word_sents.csv","r")))

y_start = data[0]
y_end = data[1]
x = data[2]

plt.figure(1)
p1, = plt.plot(x,y_start)
plt.xlabel("distance from start", fontsize = 14)
plt.ylabel("information content", fontsize = 14)

plt.figure(2)
p2, = plt.plot(x,y_end)
plt.xlabel("distance from end", fontsize = 14)
plt.ylabel("information content", fontsize = 14)
plt.title("Average Information Content for given sentence positons", fontsize = 16)
plt.show()
```
