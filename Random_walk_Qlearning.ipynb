#Using Q-learning for the random walk algorithm

```python
import random_walk
import matplotlib.pyplot as plt
import pdb

q,q_table = random_walk.main();
```

```python
%matplotlib inline

t_s2_a1 = []
q_s2_a1 = []

t_s3_a1 = []
q_s3_a1 = []

t_s3_a2 = list()
q_s3_a2 = list()

t_s4_a2 = list()
q_s4_a2 = list()

q_s2_a1.extend([q_table[key] for key in q_table if key[0]=='s_2' and key[1]=='a_1'])
t_s2_a1.extend([key[2] for key in q_table if key[0]=='s_2' and key[1]=='a_1'])

q_s3_a1.extend([q_table[key] for key in q_table if key[0]=='s_3' and key[1]=='a_1'])
t_s3_a1.extend([key[2] for key in q_table if key[0]=='s_3' and key[1]=='a_1'])
#y = [q_table[key] for key in q_table if key[0]=='s_3' and key[1] == 'a_1']

q_s3_a2.extend([q_table[key] for key in q_table if key[0]=='s_3' and key[1]=='a_2'])
t_s3_a2.extend([key[2] for key in q_table if key[0]=='s_3' and key[1]=='a_2'])

q_s4_a2.extend([q_table[key] for key in q_table if key[0]=='s_4' and key[1]=='a_1'])
t_s4_a2.extend([key[2] for key in q_table if key[0]=='s_4' and key[1]=='a_1'])

plt.plot(t_s2_a1, q_s2_a1, '*')
plt.title('Convergence of Q-values in State 2')
plt.xlabel('Time (time-steps)')
plt.ylabel('Q-values')
plt.show()

plt.plot(t_s3_a1, q_s3_a1, '*')
plt.title('Convergence of Q-values in State 3, a1')
plt.xlabel('Time (time-steps)')
plt.ylabel('Q-values')
plt.show()

plt.plot(t_s3_a2, q_s3_a2, '*')
plt.title('Convergence of Q-values in State 3, a2')
plt.xlabel('Time (time-steps)')
plt.ylabel('Q-values')
plt.show()
```
